## **Recommended System (Content-based filtering)**

### 1. **Import Libraries**

```python
import nltk
```

### 2. **Config**

```python
stopwords = nltk.corpus.stopwords.words("english")
nltk.download("stopwords")
```

### 3. **Define Util Functions**

```python
# function to return list of keywords from the input text
def get_keywords(input_text):
    if input_text is np.NaN:
        return []

    input_text = input_text.lower()

    input_text = ''.join(
        [char for char in input_text if char.isalpha() or char.isspace()])

    return [ele for ele in input_text.split() if ele not in stopwords]


# funtions to return the jaccard similarity between two lists
def jaccard_similarity(x, y):
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))

    return intersection_cardinality/float(union_cardinality)
```

### 4. **Implmenting Core**

```python
class ContentBasedFiltering:
    WEIGHT_NAME = 0
    WEIGHT_TYPE = 0
    WEIGHT_GENRE = 1

    def __init__(self, ctb_user_rating_merged_df, ctb_anime_df):
        self.user_rating_merged_df = ctb_user_rating_merged_df.copy()
        self.anime_df = ctb_anime_df.copy()

    def suggest(self, n, picked_anime_id, debug=False):
        if self.user_rating_merged_df[self.user_rating_merged_df['anime_id'] == picked_anime_id].empty:
            raise ValueError(
                'Not found picked anime id ({picked_anime_id}) in self.user_rating_merged_df')

        picked_anime_keywords = {
            'name': get_keywords(self.user_rating_merged_df[self.user_rating_merged_df['anime_id'] == picked_anime_id]['name'].tolist()[0]),
            'type': get_keywords(self.user_rating_merged_df[self.user_rating_merged_df['anime_id'] == picked_anime_id]['type'].tolist()[0]),
            'genre': get_keywords(self.user_rating_merged_df[self.user_rating_merged_df['anime_id'] == picked_anime_id]['genre'].tolist()[0])
        }

        if debug:
            print(
                f'{PRINT_COLOR_CODE.FAIL}{ContentBasedFiltering.WEIGHT_NAME} Selected Anime Keywords (name) : {PRINT_COLOR_CODE.WARNING}{picked_anime_keywords["name"]}{PRINT_COLOR_CODE.ENDC}\n'
            )
            print(
                f'{PRINT_COLOR_CODE.FAIL}{ContentBasedFiltering.WEIGHT_TYPE} Selected Anime Keywords (type) : {PRINT_COLOR_CODE.WARNING}{picked_anime_keywords["type"]}{PRINT_COLOR_CODE.ENDC}\n'
            )
            print(
                f'{PRINT_COLOR_CODE.FAIL}{ContentBasedFiltering.WEIGHT_GENRE} Selected Anime Keywords (genre) : {PRINT_COLOR_CODE.WARNING}{picked_anime_keywords["genre"]}{PRINT_COLOR_CODE.ENDC}\n'
            )

        self.anime_df['keywords_name'] = self.anime_df.apply(
            lambda x: get_keywords(x['name']), axis=1)
        self.anime_df['keywords_type'] = self.anime_df.apply(
            lambda x: get_keywords(x['type']), axis=1)
        self.anime_df['keywords_genre'] = self.anime_df.apply(
            lambda x: get_keywords(x['genre']), axis=1)

        self.anime_df['jaccard_similarity'] = self.anime_df.apply(
            lambda x:
            (ContentBasedFiltering.WEIGHT_NAME) *
            jaccard_similarity(
                picked_anime_keywords['name'], x['keywords_name']
            )

            + (ContentBasedFiltering.WEIGHT_TYPE) *
            jaccard_similarity(
                picked_anime_keywords['type'], x['keywords_type']
            )

            + (ContentBasedFiltering.WEIGHT_GENRE) *
            jaccard_similarity(
                picked_anime_keywords['genre'], x['keywords_genre']
            ),
            axis=1)

        return self.anime_df[self.anime_df['anime_id'] != picked_anime_id].sort_values(by='average_rating', ascending=False).sort_values(by='jaccard_similarity', ascending=False).head(n)
```

### 5. **Evaluation**

```python
anime_min_threshold = 1000
anime_suggestion_size = 10
```

```python
user_selection_size = 20

selected_user_ids = np.random.choice(rating_merged_df.groupby('user_id').filter(
    lambda x: len(x) >= anime_min_threshold)['user_id'].unique(), size=user_selection_size
).tolist()
print(f'{PRINT_COLOR_CODE.WARNING}selected user ids (for evaluation) : {PRINT_COLOR_CODE.CYAN}{selected_user_ids}{PRINT_COLOR_CODE.ENDC}')
```

#### 5.1. **Evaluation** using **Top-K Precision**

```python
k = anime_suggestion_size

print(
    f'Evaluation using {PRINT_COLOR_CODE.BOLD}{PRINT_COLOR_CODE.GREEN}Top-{k} Precision{PRINT_COLOR_CODE.ENDC}'
)
```

```python
avg_top_k_score = 0
for selected_user_id in selected_user_ids:
    selected_user_rating_merged_df = rating_merged_df[rating_merged_df['user_id']
                                                      == selected_user_id]
    selected_user_rating_merged_df.reset_index(drop=True, inplace=True)

    selected_anime_id = np.random.choice(
        selected_user_rating_merged_df['anime_id'].unique(), size=1)[0]

    print(
        f'\nTest by having user_id:{selected_user_id} pick anime_id:{selected_anime_id}')
    user_avg_top_k_score = 0

    ctb_suggestion_df = ContentBasedFiltering(
        selected_user_rating_merged_df, anime_df
    ).suggest(n=anime_suggestion_size, picked_anime_id=selected_anime_id)

    top_k_score = len(set.intersection(
        *[set(ctb_suggestion_df['anime_id'].head(k)), set(selected_user_rating_merged_df['anime_id'])]))/float(k)

    print(
        f'top_k_score : {top_k_score:.4f} (k = {k}, Suggest {len(ctb_suggestion_df)} animes, Answer {len(selected_user_rating_merged_df)-1} animes)'
    )

    avg_top_k_score += top_k_score

avg_top_k_score /= len(selected_user_ids)
print(
    f'\n{PRINT_COLOR_CODE.WARNING}avg_top_k_score : {PRINT_COLOR_CODE.GREEN}{avg_top_k_score}{PRINT_COLOR_CODE.ENDC}'
)
```

#### 5.2. **Evaluation** using **AUC**

```python
avg_auc_score = 0
for selected_user_id in selected_user_ids:
    selected_user_rating_merged_df = rating_merged_df[rating_merged_df['user_id']
                                                      == selected_user_id]
    selected_user_rating_merged_df.reset_index(drop=True, inplace=True)

    selected_anime_id = np.random.choice(
        selected_user_rating_merged_df['anime_id'].unique(), size=1)[0]

    print(
        f'\nTest by having user_id:{selected_user_id} pick anime_id:{selected_anime_id}')
    user_avg_auc_score = 0

    ctb_suggestion_df = ContentBasedFiltering(
        selected_user_rating_merged_df, anime_df
    ).suggest(n=anime_suggestion_size, picked_anime_id=selected_anime_id)

    ctb_suggestion_actual_interect_df = ctb_suggestion_df[ctb_suggestion_df['anime_id'].isin(
        selected_user_rating_merged_df['anime_id'])].copy()
    ctb_suggestion_not_interect_df = ctb_suggestion_df[~ctb_suggestion_df['anime_id'].isin(
        selected_user_rating_merged_df['anime_id'])].copy()

    auc_score = 0

    n = len(ctb_suggestion_actual_interect_df) * \
        len(ctb_suggestion_not_interect_df)
    if n == 0:
        auc_score = 0
        print(
            f'auc_score : {auc_score:.4f} (n = {n}, n\' = -, n\" = -) (actual_interect : {len(ctb_suggestion_actual_interect_df)}, not_interect : {len(ctb_suggestion_not_interect_df)}) (Suggest {len(ctb_suggestion_df)} animes, Answer {len(selected_user_rating_merged_df)-1} animes)'
        )
    else:
        ctb_suggestion_actual_interect_df['n_prime'] = ctb_suggestion_actual_interect_df.apply(
            lambda x: len(ctb_suggestion_not_interect_df[x['jaccard_similarity'] > ctb_suggestion_not_interect_df['jaccard_similarity']]), axis=1)
        ctb_suggestion_actual_interect_df['n_prime_2'] = ctb_suggestion_actual_interect_df.apply(
            lambda x: len(ctb_suggestion_not_interect_df[x['jaccard_similarity'] == ctb_suggestion_not_interect_df['jaccard_similarity']]), axis=1)

        n_prime = ctb_suggestion_actual_interect_df['n_prime'].sum()
        n_prime_2 = ctb_suggestion_actual_interect_df['n_prime_2'].sum()

        auc_score = (n_prime + (0.5 * n_prime_2)) / n
        print(
            f'auc_score : {auc_score:.4f} (n = {n}, n\' = {n_prime} , n\" = {n_prime_2}) (actual_interect : {len(ctb_suggestion_actual_interect_df)}, not_interect : {len(ctb_suggestion_not_interect_df)}) (Suggest {len(ctb_suggestion_df)} animes, Answer {len(selected_user_rating_merged_df)-1} animes)'
        )

    avg_auc_score += auc_score

avg_auc_score /= len(selected_user_ids)
print(
    f'\n{PRINT_COLOR_CODE.WARNING}avg_auc_score : {PRINT_COLOR_CODE.GREEN}{avg_auc_score}{PRINT_COLOR_CODE.ENDC}'
)
```
